# Importando dados e análisando

```python
dados = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/latest/owid-covid-latest.csv")
dados.head()
```

```python
dados.info()
```

```python
dados.isnull().sum()
```

Vou usar o critério de que se alguma coluna de dados tiver valores nulos maior
que 200 vou deletar

```python
dados_nulos_drop = dados.isnull().sum() > 200
dados_nulos_drop = dados_nulos_drop.loc[dados_nulos_drop == True]
for colunas in dados_nulos_drop.index:
  dados.drop(colunas,axis=1,inplace=True)
```

```python
dados.head()
```

```python
# Substituindo pela media de cada coluna
colunas_numericas = dados.info()
colunas_numericas
```

```python
numericos = ['int64','float64']
colunas_numericas = dados.select_dtypes(numericos)
colunas_numericas
```

*texto em itálico*# Identifica outliers

```python
primeiro_quartil = colunas_numericas.quantile(0.25)
terceiro_quartil = colunas_numericas.quantile(0.75)
distancia_interquartil = terceiro_quartil - primeiro_quartil
limite_superior = primeiro_quartil + 1.5 * distancia_interquartil
limite_inferior = primeiro_quartil - 1.5 * distancia_interquartil
```

```python
# limite superior
limite_superior = colunas_numericas > limite_superior
outlietrs = limite_superior == True
outlietrs
```

```python
limite_inferior = colunas_numericas < limite_inferior
(limite_inferior == True).sum()
```
