```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/latest/owid-covid-latest.csv")
dados.head()
```

```python
dados.isnull().sum()
```

```python
dados_nulos_drop = dados.isnull().sum() > 200
dados_nulos_drop = dados_nulos_drop.loc[dados_nulos_drop == True]
for colunas in dados_nulos_drop.index:
  dados.drop(colunas,axis=1,inplace=True)
```

```python
dados_nulos = dados.isnull().sum() == 0
dados_nulos = dados_nulos.loc[dados_nulos == False]
colunas_numericas = dados.select_dtypes('float') 
media_de_cada_coluna = colunas_numericas.mean()
media_de_cada_coluna = media_de_cada_coluna.round(3)
for colunas in dados_nulos.index:
  dados.fillna(media_de_cada_coluna,inplace=True)
```
