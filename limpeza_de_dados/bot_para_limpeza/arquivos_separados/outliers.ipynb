```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
dados
```

#Z-score
O z-score nos diz o quanto cada valor em nossa distribuiÃ§Ã£o se distancia da mÃ©dia em termos de desvio padrÃ£o.

```python
z_score = dados['medv'] - np.mean(dados['medv']) / np.std(dados['medv'])
print(z_score)
print('-'*90)
print(np.abs(z_score))
```

```python
z_score = np.abs(stats.zscore(dados['medv']))
print(z_score)
```
