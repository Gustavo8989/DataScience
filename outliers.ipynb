```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
dados
```

#Z-score
O z-score nos diz o quanto cada valor em nossa distribuição se distancia da média em termos de desvio padrão.

```python
z_score = dados['medv'] - np.mean(dados['medv']) / np.std(dados['medv'])
print(z_score)
print('-'*90)
print(np.abs(z_score))
```

```python
z_score = np.abs(stats.zscore(dados['medv']))
print(z_score)
```

```python
print(np.array(z_score))
```

```python
limite_thershold = 3
np.where(z_score > 3)
```

# IQR (intervalo interquartil)

a fim de avaliar o grau de espalhamento de dados (dispersão) em torno da medida de centralidade


# Medida de centralidade
Medida de centralidade são usadas para representar toda uma lista de observações com um único valor




```python
q1 = np.percentile(dados['medv'],25)
q3 = np.percentile(dados['medv'],75)
iqr = q1 - q3
print(iqr)
```

```python
print({q3+1.5*iqr})
print(q1+1.5*iqr)
print('-'*89)
print(dados['medv'])
```

```python
acima = dados['medv'] >= (q3+1.5*iqr)
print(acima)
abaixo = dados['medv'] <= (q1+1.5*iqr)
print('-'*90)
print(abaixo)
```

```python
print(np.where(acima))
print('-'*90)
print(np.where(abaixo))
```
